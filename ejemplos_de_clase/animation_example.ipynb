### Import libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import random
from itertools import count
from matplotlib.animation import FuncAnimation
```

### Enable animations in Jupyter

```python
%matplotlib nbagg
```

### Simple example with random data

```python
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    plt.cla() # clear previous axis
    plt.plot(x_vals, y_vals)

ani = FuncAnimation(plt.gcf(), animate, interval=1000)    
    
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []

index = count()
data = pd.read_csv('./data/animation_data/data.csv')

def animate(i):
    
    x = data['x_value'][0:i]
    y1 = data['total_1'][0:i]
    y2 = data['total_2'][0:i]

    plt.cla()

    plt.plot(x, y1, label='Channel 1')
    plt.plot(x, y2, label='Channel 2')

    plt.legend(loc='upper left')
    plt.tight_layout()
    i=+1


ani = FuncAnimation(plt.gcf(), animate, interval=1000)

plt.tight_layout()
plt.show()
```
